## Transformation 

```python
#merging company tweet, and tweet dfs on "tweet_id"
merged_df = pd.merge(tweet_id, tweet_df, on="tweet_id")
merged_df.head()
```

```python
merged_df.ticker_symbol.unique()
```

```python
# Since we have 2 google, removing all the googl ticker 
merged_df = merged_df[merged_df.ticker_symbol !="GOOGL"]
```

```python
merged_df.ticker_symbol.unique()
```

```python
#converting unix epoch to date format similar
merged_df["post_date"] = pd.to_datetime(merged_df["post_date"], unit='s').dt.strftime('%d-%m-%Y')
merged_df.tail()
```

```python
#checking values before dropping null values
merged_df.count()
```

```python
#dropping null values
merged_df = merged_df.dropna()
```

```python
#checking values/count after dropping null values
merged_df.count()
```

```python
# Checking start and end date of the tweet data 
merged_df = merged_df.sort_values(by='post_date')
merged_df.head()
```

```python
merged_df.tail()
```

```python
# Double check nulls
merged_df.isnull().sum()
```

```python
# remane post_date 
merged_df = merged_df.rename(columns={'post_date': 'Date'}, errors='raise')
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['writer', 'body'])
merged_df.head()
```

##tweet activity df for each tickers

Apple

```python
# columns
cols = ["Date","tweet_count"]
```

```python
AAPL = merged_df[(merged_df['ticker_symbol']=="AAPL")]
AAPL 
```

```python
#creading df of tweets per day to merge with stock dfs
aapl_df = pd.DataFrame(AAPL.Date.value_counts()).reset_index()
#setting the appropriate columns
aapl_df.columns=cols

aapl_df
```

Amazon

```python
AMZN = merged_df[(merged_df["ticker_symbol"]=="AMZN")]

amzn_df = pd.DataFrame(AMZN.Date.value_counts()).reset_index()
amzn_df.columns=cols

amzn_df
```

Google


```python
GOOG = merged_df[(merged_df["ticker_symbol"]=="GOOG")]

goog_df = pd.DataFrame(GOOG.Date.value_counts()).reset_index()
goog_df.columns=cols

goog_df
```

Microsoft

```python
MSFT = merged_df[(merged_df["ticker_symbol"]=="MSFT")]

msft_df = pd.DataFrame(MSFT.Date.value_counts()).reset_index()
msft_df.columns=cols

msft_df
```

Tesla

```python
TSLA = merged_df[(merged_df["ticker_symbol"]=="TSLA")]

tsla_df = pd.DataFrame(TSLA.Date.value_counts()).reset_index()
tsla_df.columns=cols

tsla_df
```

merging tweet activity to original

```python
aapl_merged = pd.merge(AAPL, aapl_df, on='Date')
aapl_merged.head()
```

```python
amzn_merged = pd.merge(AMZN, amzn_df, on='Date')
amzn_merged.head()
```

```python
goog_merged = pd.merge(GOOG, goog_df, on='Date')
goog_merged.head()
```

```python
msft_merged = pd.merge(MSFT, msft_df, on='Date')
msft_merged.head()
```

```python
tsla_merged = pd.merge(TSLA, tsla_df, on='Date')
tsla_merged.head()
```

## final merge


Apple


```python
df_aapl_f = df_aapl
df_aapl_f
```

```python
df_aapl_f['Date'] = pd.to_datetime(df_aapl_f.index)
df_aapl_f["Date"] = pd.to_datetime(df_aapl_f["Date"], unit='s').dt.strftime('%d-%m-%Y')

df_aapl_f.reset_index(drop=True, inplace=True)
df_aapl_f
```

```python
final_apple = pd.merge(df_aapl_f, aapl_df, on='Date')
final_apple.head()
```

```python
final_apple['price_action'] = final_apple['Close'] - final_apple['Open']

final_apple.tail()
```

```python
final_apple['Ticker'] = "AAPL"
final_apple.head()
```
